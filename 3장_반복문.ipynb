## 반복문

같은 작업을 반복적으로 해야할 때, 반복문을 사용한다면 작업의 양이 줄어들 뿐만 아니라 보기에도 깔끔해지게 된다.
반복문에는 크게 for 와 while 두 가지가 있다.

```python
fruit=["사과","바나나","멜론","귤","딸기"]
print("나는 "+fruit[0]+"을(를) 먹습니다.")
print("나는 "+fruit[1]+"을(를) 먹습니다.")
print("나는 "+fruit[2]+"을(를) 먹습니다.")
print("나는 "+fruit[3]+"을(를) 먹습니다.")
print("나는 "+fruit[4]+"을(를) 먹습니다.")
```

```python
for i in fruit:
    print("나는 "+i+"을(를) 먹습니다.")
```

### For

for는 list에 있는 항목들을 한 번씩만 훑어보기 위해 사용한다.

for의 구조는 다음과 같다

for variable in list:

    contents

if절과 같이 for 즉 반복적으로 행할 내용들은 for에 속하는 것임을 나타내기 위해서 indentation(공백)을 만들어 둔다.

다음과 같이 list 대신 string을 넣으면 한 글자(char)씩 반복하며 진행한다.

```python
country="Republic Of Korea"
for ch in country:
    if ch.isupper():
        print(ch)
```

### Range

for 반복문을 사용할 때 in 뒷 부분에 list나 string과 같이 반복가능한 객체(iterable)을 사용해서 직접 값에 접근을 하기도 하지만 range를 사용해서 인덱스로 접근을 하기도 한다.

range는 range((start),stop,(step))으로 구성되어 있다.

start(필요시)부터 stop(포함되어야 함)까지 step(필요시)씩 연속적인 정수 객체를 만들어준다. 

```python
for num in range(10):
    print(num, end=" ")
```

```python
for num in range(5,10):
    print(num, end=" ")
```

```python
for num in range(0,10,2):
    print(num,end=" ")
```

range함수로 만들어진 값은 list와 비슷해보이지만 range라는 객체를 가지는 것이므로 list에서 사용하는 함수를 사용할 수는 없다.

```python
a=range(7)
print(type(a))
```

인덱스로 접근하여 사용하려 할 경우 다음과 같이 사용한다.

```python
metals=["Li","Na","K"]
weights=[6.941,22.98976928,39.0983]

for i in range(len(metals)):
    print(metals[i],weights[i])
```

### enumerate

이 점을 보완해서 값과 인덱스를 동시에 접근할 수 있도록 해주는 함수이다.

enumerate(iterable)로 사용을 하며 enumerate 객체를 반환을 하며 인덱스와 값을 쌍으로 가지는 튜플값을 하나의 원소로 가진다.

(0,iterable[0]),(1,iterable[1]),....

```python
values=[8,-4,23,14,9]
for i,v in enumerate(values):
    print(i,v)
```

### 이중반복문

이중반복문이랑 반복문안에 반복문을 사용하는 것으로 주로 행렬을 나타내고자 할 때, 서로 연관있는 두 개의 반복가능한 객체를 동시에 사용하고자 할 때 사용한다.

```python
first=["1","2","3"]
second=["1.0","2.0","3.0"]
for i in first:
    for j in second:
        print(i+"번째 중 "+j+"번째")
    print("----------------")
```

위처럼 두 개의 반복문을 사용할 때 서로 각각의 variable을 사용해야 하며 first의 한 값당 second가 한 번씩 실행됨을 알 수 있다.

또한 들여쓰기에 따라서 실행되는 횟수가 달라지는데 print(i+"번째 중 "+j+"번째")는 i 와 j 둘다 속해있는 부분이라 총 9번 실행됨을 알 수 있지만

print("----------------")는 i에만 해당하는 부분이므로 3번만 실행됨을 알 수 있다.

## while

while 반복문은 내가 얼마나 반복문을 사용해야할지 정확히 모를 때 사용하거나 중복적인 인덱스 혹은 값을 접근해야 할 경우에 사용한다.

while의 구조는 다음과 같다.

while condition:

    contents
    
while은 for와 달리 순차적으로 진행을 하지 않고 condition이 참인 경우에만 while 내의 내용을 반복한다. 이 말은 condition이 거짓이 될 때까지 계속 반복한다는 뜻이다. 따라서 처음에는 condition이 참이여야 하지만 contents를 진행하면서 condition이 거짓이 되도록 만들어야 한다. 그렇지 않으면 무한반복에 빠지게 된다.
만약 무한반복에 빠지게 된다면 ctrl+c 혹은 쥬피터를 사용한다면 위의 정지버튼을 누르면 된다.

```python
apple=3
while apple>0:
    print(apple)
    apple-=1
```

```python
apple=3
while apple>0:
    print(apple)
```

반복문을 더 효율적으로 활용하기 위해서 break와 continue를 사용할 수 있다.

### break

반복문 진행중에 이미 원하는 바를 이뤄서 뒤의 진행이 필요가 없을 수 있다. 따라서 더 효율적으로 사용하기 위해 반복문이 한참 진행중이라도 일정 조건을 만나면 반복문이 끝나게 만들어주는 것이 break다.
break는 다음과 같이 사용한다.

```python
for i in range(15):
    if i==10:
        break
    else:
        print(i,end=" ")
```

### continue

break는 일정 조건을 만나면 반복문을 깨뜨리지만 continue는 일정 조건에서만 반복문이 진행되지 않도록 한다. 반복문 내에서 continue를 만난다면 아직 반복문 내에 남은 내용이 있더라도 작동하지 않고 다음 값으로 넘어간다.
continue는 다음과 같이 사용한다.

```python
for i in range(16):
    if i%5==0:
        continue
    print(i,end=" ")
```

### One-line for

python만의 장점으로 간단한 반복문은 한 줄로 간단하게 표현할 수 있다. 우선 다음과 같은 반복문 예제가 있다.

```python
students = [1, 2, 3, 4, 5]
for i in range(len(students)):
    students[i] = students[i]+100
print(students)
```

위와 동일한 과정을 수행할 수 있는 반복문을 한줄로 표현할 수 있다.

```python
students = [101+i for i in range(5)]
print(students)
```

```python
students = [0 for _ in range(5)]
print(students)

from pprint import pprint
dp = [[0 for _ in range(10)] for _ in range(10)]
pprint(dp)
```
